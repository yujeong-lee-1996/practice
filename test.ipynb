```python
from gensim.models import Word2Vec

# 모델 로드
model_path = "model/my_word2vec.model"  # 경로 맞춰서 수정
model = Word2Vec.load(model_path)

```

```python
# 모든 단어 리스트 출력 (vocabulary)
vocab_words = model.wv.index_to_key  # 또는 model.wv.key_to_index.keys()
print(vocab_words) 
```

```python
# 예시 벡터 연산
result = model.wv.most_similar(positive=["힙합", "감성"], negative=["애쉬"], topn=1)
print("결과:", result)
```

```python
# 벡터 유사도
sim = model.wv.similarity("강아지", "고양이")
print(sim)


```

```python

# 어울리지 않는 단어
odd = model.wv.doesnt_match(["삼성", "LG", "애플", "치킨"]) 
print(odd)
```

```python
# 단어 벡터 얻기
vec = model.wv.get_vector("사랑")

# 벡터로 유사한 단어 찾기
model.wv.similar_by_vector(vec, topn=5)
```
